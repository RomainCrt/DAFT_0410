#### Display the trending Python repositories in GitHub.

The steps to solve this problem is similar to the previous one except that you need to find out the repository names instead of developer names.

````python
# This is the url you will scrape in this exercise
url = 'https://github.com/trending/python?since=daily'
````

````python
# your code here

html = requests.get(url).content
soup = BeautifulSoup(html, "lxml")
Lst = soup.find_all('h2', {'class': 'h3 lh-condensed'})
LstName = []
for i in Lst:
    MyStr = i('a')[0].text.strip()
    Pos= MyStr.find('/')
    MyStr = MyStr[:Pos].strip() + '/' + MyStr[Pos +1:].strip()
    LstName.append(MyStr)
print(LstName)
````

#### Find the number of titles that have changed in the United States Code since its last release point.

````python
# This is the url you will scrape in this exercise
url = 'http://uscode.house.gov/download/download.shtml'
````

````python
import requests
from bs4 import BeautifulSoup

response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
#print(soup.prettify())
titles = soup.find_all('div', {'class': 'usctitlechanged'})
for i in titles:
    print(i.get_text().strip())

    
````

####  Display the 20 latest earthquakes info (date, time, latitude, longitude and region name) by the EMSC as a pandas dataframe.

````python
# This is the url you will scrape in this exercise
url = 'https://www.emsc-csem.org/Earthquake/'
````

````python
# your code here
import pandas as pd

response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
hq = []
for elem in soup.find_all(class_='ligne1 normal'):
   
    hq_l = elem.text[10:-1]
    hq.append(elem.text[10:-1])
    print((elem.text))
    
# Display the first 20 elements in a pandas dataframe
df = pd.DataFrame({'hq' : hq[:20]})
print(df)
````

````python
import requests
from bs4 import BeautifulSoup

url ='https://www.imdb.com/chart/top'
response = requests.get(url).content
sp = BeautifulSoup(response, 'html.parser')


print(sp)
````

````python
selector = '#main > div > span > div > div > div.lister > table > tbody > tr:nth-child(2) > td.titleColumn > a'
````

````python
sp.select(selector)[0]
````

````python
sp.select(selector)[0].text
````

````python
sp.select(selector)[0].get_text()
````

````python
selector2 = 'td.titleColumn > a'
sp.select(selector2)
print(len(sp.select(selector2)))
print(sp.select(selector2))
````

````python
movie_tittles = []
for link in sp.select(selector2):
    movie_tittles.append(link.get_text())

selector3 = "td.titleColumn > span"
movie_year = []
for link in sp.select(selector3):
    movie_year.append(int(link.get_text()[1:-1]))
    
````

````python
movie_df = pd.DataFrame({'movie_tittlES' : movie_tittles,'year' : movie_year})
````
