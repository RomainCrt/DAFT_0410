```python
import pandas as pd
df = pd.read_csv('example_data_cleaning.csv', sep=',')
```

```python
df
```

```python
#Removing columns Unnamed: 7  & Unnamed: 8    (Unnamed: 8 is equal to TransactionID with a rows +1 offset)
df.drop("Unnamed: 7",axis=1,inplace=True)
df.drop("Unnamed: 8",axis=1,inplace=True)
```

```python
#Removing the last line
#df.drop(df['TransactionID'].isna(),inplace=True, axis=1)
df = df.drop(df[df['TransactionID'].isna()].index)
```

```python
from sklearn.impute import SimpleImputer, KNNImputer
```

```python
#Testing NaN values
display(df.loc[df["TransactionID"].isna()])
display(df.loc[df["ClientID"].isna()])
display(df.loc[df["BirthYear"].isna()])
display(df.loc[df["Amount"].isna()])
display(df.loc[df["Profession"].isna()])
display(df.loc[df["Department"].isna()])
display(df.loc[df["Risk"].isna()])
```

BirthYear and Profession need impute

```python
#BirthYear is numeric
imputer = KNNImputer(n_neighbors=5)
df["BirthYear"] = imputer.fit_transform(df[["BirthYear"]]).astype(int)
```

```python
#Profession is string
imputer = SimpleImputer(strategy='most_frequent')
df["Profession"] = imputer.fit_transform(df[["Profession"]])
```

```python
#Encoding Risk category
df["NRisk"] = df["Risk"].map({"High": 3, "Medium" : 2, "Low" : 1})
```

```python
#Profession looks weird 
df["Profession"].unique()
```

```python
#Normalisation of Profession  1. Lower all
df["Profession"] = df["Profession"].str.lower()
df.loc[df["Profession"] == 'etudient' ,"Profession"] = "student"
```
